```python
# Install required libraries
# pip install torch torchvision transformers pillow

import torch
from transformers import AutoImageProcessor, SwinForImageClassification
from PIL import Image
import requests

# 1. Load Pre-trained Model and Processor
model_name = "microsoft/swin-tiny-patch4-window7-224"
# The processor handles image resizing and normalization
image_processor = AutoImageProcessor.from_pretrained(model_name)
model = SwinForImageClassification.from_pretrained(model_name)

# 2. Prepare an Image
#url = "http://images.cocodataset.org"
#image = Image.open(requests.get(url, stream=True).raw)

```

```python
image = Image.open("Huskiesatrest.jpg")

# 3. Preprocess the image and convert to tensor
inputs = image_processor(images=image, return_tensors="pt")

# 4. Forward Pass
with torch.no_grad():
    outputs = model(**inputs)
    logits = outputs.logits

# 5. Get Prediction
predicted_class_idx = logits.argmax(-1).item()
print("Predicted class:", model.config.id2label[predicted_class_idx])
# Expected output: Predicted class: Egyptian cat
```
